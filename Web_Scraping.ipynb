```python
# import selium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Make browser headless
chrome_options = Options()
chrome_options.add_argument('--headless')  # Run Chrome in headless mode
chrome_options.add_argument('--no-sandbox') # Disable sandboxing, necessary in Colab
chrome_options.add_argument('--disable-dev-shm-usage') # Prevent shared memory issues

# Set up the WebDriver
driver = webdriver.Chrome(options=chrome_options)
driver.command_executor.set_timeout(1000)
```

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup

union_dfs = pd.DataFrame()

for i in range(2017,2018): # (2017, 2023)
  print("started for loop")
  URL = "https://www.basketball-reference.com/leagues/NBA_{}_transactions.html".format(i) #"https://realpython.github.io/fake-jobs/"
  #page = requests.get(URL)
  driver.get(URL.format(i))
  soup = BeautifulSoup(driver.page_source, 'html.parser')
  #print(soup)

  job_elements = soup.find_all("ul", class_="page_index")

  #print(type(job_elements))
  dates = []
  moves = []
  for i in job_elements:

    #print(i.find_all("li"))
    i = i.find_all("li") # Added: 1/20
    print("I", i)
    for j in i:
      #print("J", j)
      k = j.find_all("p") # Added: 1/20
      date = j.find("span")
      print("K",k)
      for move in k:  # Added: 1/20
        #move = m.find("p") #.split("<"))
        print("MOVE", move.text.strip())

        if type(move) != type(0):
          dates.append(date.text.strip())
          #print(date.text.strip())
          moves.append(move.text.strip())
          #print(move.text.strip())

  driver.quit()
  df = pd.DataFrame(moves,columns=['Action'])
  df['Dates'] = dates
  print("DATES", dates)
```
