```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
import shutil

def create_driver():
    chrome_path = shutil.which("google-chrome")

    chrome_options = Options()
    chrome_options.binary_location = chrome_path
    chrome_options.add_argument("--headless=new")  # if this breaks, use "--headless"
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")

    service = Service("/usr/local/bin/chromedriver")  # same path you used before

    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from urllib.parse import quote_plus

def scrape_zara_search(query, section="WOMAN", country="us", lang="en",
                       max_scrolls=5, scroll_pause=1.5):
    base_url = f"https://www.zara.com/{country}/{lang}/search"
    q = quote_plus(query)
    url = f"{base_url}?searchTerm={q}&section={section}"

    #driver = create_driver()
    print("Opening:", url)
    driver.get(url)

    wait = WebDriverWait(driver, 15)
    try:
        container = wait.until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, "ul.product-grid__product-list")
            )
        )
    except Exception as e:
        print("Could not find product grid:", e)
        driver.quit()
        return pd.DataFrame()

    last_height = driver.execute_script("return document.body.scrollHeight")
    for _ in range(max_scrolls):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    container = driver.find_element(By.CSS_SELECTOR, "ul.product-grid__product-list")
    products = container.find_elements(By.CSS_SELECTOR, "li.product-grid-product")
    print("Found products:", len(products))

    data = {"name": [], "price": [], "discount": [], "url": []}

    for p in products:
        try:
            title_el = p.find_element(
                By.CSS_SELECTOR,
                'a.product-link._item.product-grid-product-info__name.link'
            )
            name = title_el.text.strip()
            href = title_el.get_attribute("href")

            try:
                price_el = p.find_element(
                    By.CSS_SELECTOR,
                    "span.money-amount__main"
                )
                price = price_el.text.strip()
            except Exception:
                price = None

            try:
                disc_el = p.find_element(
                    By.CSS_SELECTOR,
                    "span.price-current__discount-percentage"
                )
                discount = disc_el.text.strip()
            except Exception:
                discount = None

            data["name"].append(name)
            data["price"].append(price)
            data["discount"].append(discount)
            data["url"].append(href)
        except Exception as e:
            print("Skipped one product due to error:", e)
            continue

    driver.quit()
    return pd.DataFrame(data)
```

```python
from selenium.webdriver.common.by import By
import time
import pandas as pd

def scrape_zara_search_simple(query, section="WOMAN", country="us", lang="en",
                              max_scrolls=5, scroll_pause=1.5):
    from urllib.parse import quote_plus

    base_url = f"https://www.zara.com/{country}/{lang}/search"
    q = quote_plus(query)
    url = f"{base_url}?searchTerm={q}&section={section}"

    driver = create_driver()
    print("Opening:", url)
    driver.get(url)

    # Let JS do its thing
    time.sleep(5)

    # Scroll a bit to trigger lazy loading
    last_height = driver.execute_script("return document.body.scrollHeight")
    for _ in range(max_scrolls):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # DEBUG: check if the product class appears anywhere in the HTML
    page = driver.page_source
    print("Contains 'product-grid-product' in HTML?:",
          "product-grid-product" in page)

    # Try to grab products directly by XPATH
    products = driver.find_elements(
        By.XPATH, '//li[contains(@class, "product-grid-product")]'
    )
    print("Found products:", len(products))

    data = {"name": [], "price": [], "discount": [], "url": []}

    for p in products:
        try:
            # Name + URL
            title_el = p.find_element(
                By.XPATH,
                './/a[contains(@class, "product-link") and contains(@class, "product-grid-product-info__name")]'
            )
            name = title_el.text.strip()
            href = title_el.get_attribute("href")

            # Price
            try:
                price_el = p.find_element(
                    By.XPATH,
                    './/span[contains(@class, "money-amount__main")]'
                )
                price = price_el.text.strip()
            except Exception:
                price = None

            # Discount (if any)
            try:
                disc_el = p.find_element(
                    By.XPATH,
                    './/span[contains(@class, "price-current__discount-percentage")]'
                )
                discount = disc_el.text.strip()
            except Exception:
                discount = None

            data["name"].append(name)
            data["price"].append(price)
            data["discount"].append(discount)
            data["url"].append(href)

        except Exception as e:
            print("Skipped one product due to error:", e)
            continue

    driver.quit()
    return pd.DataFrame(data)
```

```python
def scrape_zara_category_simple(url, max_scrolls=5, scroll_pause=1.5):
    driver = create_driver()
    print("Opening:", url)
    driver.get(url)

    import time
    time.sleep(5)

    last_height = driver.execute_script("return document.body.scrollHeight")
    for _ in range(max_scrolls):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    page = driver.page_source
    print("Contains 'product-grid-product' in HTML?:",
          "product-grid-product" in page)

    from selenium.webdriver.common.by import By
    products = driver.find_elements(
        By.XPATH, '//div[contains(@class,"product-grid-product-info")]'
    )
    print("Found products:", len(products))

    import pandas as pd
    data = {"name": [], "price": [], "url": []}

    for p in products:
        try:
            name_el = p.find_element(By.XPATH, ".//h2")
            name = name_el.text.strip()

            try:
                price_el = p.find_element(
                    By.XPATH,
                    './/span[contains(@class,"money-amount__main")]'
                )
                price = price_el.text.strip()
            except Exception:
                price = None

            link_el = p.find_element(By.XPATH, ".//a")
            href = link_el.get_attribute("href")

            data["name"].append(name)
            data["price"].append(price)
            data["url"].append(href)
        except Exception as e:
            print("Skipped one product:", e)
            continue

    driver.quit()
    return pd.DataFrame(data)
```
