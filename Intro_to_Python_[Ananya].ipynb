## Read from a file

1. open file for reading
2. read data from file
3. store and analyze data

Open file and read data

```python
# Using read() vs. readlines() to read contents from a file

# choose a filename and directory
read_directory = "/content/gdrive/Shareddrives/CSC102_Files/Datasets_for_Students/"

#filename = "names.txt"
#filename = "Streaming_Services_2025.csv"
filename = "Streaming_Services_2025.txt"

# open file using read_directory and filename, and store in my_file
my_file = open(read_directory + filename, "r")


# read and store the contents using read(), then readlines()
contents = my_file.readlines()

# print the contents and its type (hint: try using repr() function)
print(repr(contents))
print(type(contents))
```

Store data

```python
# Loop through each line of contents and print each line
# Use .split() to split each line into a list and print the list
# Store the data (service name and price) in an appropriate container

service_dictionary = {}
contents.pop(0)
for line in contents: #[1:]
  print(repr(line))
  my_list = line.split()
  print(my_list)
  key = my_list[0]
  value = my_list[1]
  service_dictionary[key] = float(value)

print(service_dictionary)

```

#### Read the data into a pandas dataframe and output the first 10 rows

```python
url = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv"
df = pd.read_csv(url)
df = df.sample(frac=1, random_state=36) # random_state for reproducibility


# Output the first 10 rows of the dataframe
```

#### How many rows and columns are in the dataframe? What are the names of the columns?

#### Find the maximum rating. Output the maximum rating and the title of the movie.

```python
# Find the maximum rating
maximum_rating = df.star_rating.max()
print(maximum_rating)

# Filter the dataframe to return only the rows with the maximum rating
max_movie = df[df['star_rating'] == maximum_rating]
print(max_movie)

# Output the titles of the movies with the maximum rating
print(max_movie['title'])
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Filter the first 10 movies in the dataframe
df10 = df.iloc[0:10]

# Plot the title as the x-axis and the rating as the y-axis
plt.plot(df10["title"], df10["star_rating"])
# Rotate the x-ticks 90 degrees
plt.xticks(rotation=90)
plt.show()
```
